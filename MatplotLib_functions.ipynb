```python
# This is your given monte carlo pie solution because I did not know how yield works.
import random
import sys
def mc_pi():
    inside_unit_circle = total_points = 0
    while True:
        total_points += 1
        a, b = random.random(), random.random()
        if a ** 2 + b ** 2 < 1:
            inside_unit_circle += 1
        yield 4 * inside_unit_circle / total_points


def main(max_iterations=200):
    for i, pi in enumerate(mc_pi()):
        if i > max_iterations:
            break
    print(pi)
main(1000)
print(mc_pi())
```

```python
import numpy as np

def numpy_Pi(num_simulations):
    a, b = np.random.random(size=(2,num_simulations))#Insted of generating numbers it adds inside-circles to numpy array
    inside_circle = (np.sqrt(a**2 + b**2) < 1)
    return  4*(inside_circle.sum()/num_simulations)

num_simulations=50000
pi = numpy_Pi(num_simulations)
print(pi)
```
